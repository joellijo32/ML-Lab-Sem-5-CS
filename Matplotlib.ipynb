```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

plt.style.use('dark_background') # seaborn, ggplot
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Graph')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.plot(y, x)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Graph')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X axis');
ax.set_ylabel('Y axis')
ax.set_title('SubPlot')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title("Sin(X)")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title("Sin(X)")
# plt.grid(True)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.plot(x, y3, label='sin(x)cos(x)')
plt.xlabel('X')
plt.ylabel('sin(x)')
plt.title('Trignometric functions')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

```

# Scatter Plot


```python
np.random.seed(42)
x = np.random.randn(100)
y = 2 * x * np.random.randn(100) * 0.5

plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.8)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot')
plt.grid(True, alpha=0.1)
plt.show()
```

# Bar Charts


```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 32]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='steelblue', edgecolor='orange', linewidth=2.1)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.title('Bar Chart')
plt.grid(True, alpha=0.2, axis='y')
plt.show()
```

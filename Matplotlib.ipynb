```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

plt.style.use('default') # seaborn, ggplot
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Graph')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.plot(y, x)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Graph')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlabel('X axis');
ax.set_ylabel('Y axis')
ax.set_title('SubPlot')
plt.show()
```
